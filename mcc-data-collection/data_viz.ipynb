```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_hdf5_time(filename):
    # Load the file
    # CSV:
    df = pd.read_csv(filename)
    # .hf5:
    #df = pd.read_hdf(filename, 'df')

    # Convert the 'Time' column to float values
    df['time'] = df['time'].astype(float)

    # Plot each column against 'Time'
    for col in df.columns:
        if col not in ['time', 'start_time']:
            plt.figure(figsize=(10, 5)) # Create a new figure for each plot
            plt.plot(df['time'], df[col])
            plt.xlabel('time')
            plt.ylabel(col)
            plt.title(f'{col} vs Time')
            #plt.xlim([130, 140])
            plt.show()
            

def diagnostic_info(filename):
    # Load the file
    # CSV:
    df = pd.read_csv(filename)
    # .hf5:
    #df = pd.read_hdf(filename, 'df')

    # Calculate the number of samples (rows)
    num_samples = len(df)

    # Calculate the total time of recording (the last time value minus the first one)
    total_time = df['time'].iloc[-1] - df['time'].iloc[0]

    # Calculate the effective sampling rate (total number of samples divided by the total time)
    eff_sampling_rate = num_samples / total_time
    
    file_size = os.path.getsize(filename)
    print(f"Current file size: {file_size / 1024 / 1024:.2f} MB")   

    print(f"Number of samples: {num_samples}")
    if total_time < 60:
        print(f"Total time of recording: {round(total_time,2)} seconds")
    elif total_time < 60*60:
        print(f"Total time of recording: {round(total_time/60,2)} minutes")
    elif total_time < 60*60:
        print(f"Total time of recording: {round(total_time/(60*60),2)} hours")
    print(f"Effective sampling rate: {eff_sampling_rate} Hz")  
```

```python
filename = "data/"
filename = filename + "Hat_1_2023-05-26_18:41:55"
filename = filename + ".csv"

diagnostic_info(filename)
plot_hdf5_time(filename)
#plot_hdf5_timestamp(filename)
```
